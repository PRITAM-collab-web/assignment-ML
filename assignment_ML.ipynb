Question 2: What are the types of machine learning? Describe each with one
real-world example.

--> The primary types of machine learning are Supervised Learning, where models learn from labeled data to make predictions or classifications; Unsupervised Learning, which finds patterns in unlabeled data; Reinforcement Learning, where an agent learns through trial and error to maximize rewards; Semi-Supervised Learning, a hybrid of supervised and unsupervised with a mix of labeled and unlabeled data; and Self-Supervised Learning, which generates its own labels from the data to learn from.
Here's a description of each type with a real-world example:
1. Supervised Learning
Description:
In supervised learning, the model is trained on a labeled dataset, meaning each input example has a corresponding correct output. The goal is to learn a mapping from the input to the output so that it can predict the output for new, unseen data.
Example:
A model is trained on thousands of emails that are labeled as either "spam" or "not spam". It learns to identify patterns in spam emails, and then uses this knowledge to classify new, incoming emails as spam or legitimate.
2. Unsupervised Learning
Description:
Unsupervised learning involves training a model on unlabeled data to find inherent patterns, structures, or relationships within the data.
Example:
A retail company can use unsupervised learning to group customers into different segments based on their purchasing habits, demographics, and behaviors.
3. Reinforcement Learning
Description:
In this type of learning, a software agent learns to make decisions by performing actions in an environment and receiving feedback in the form of rewards or punishments.
Example:
An autonomous car's driving system can use reinforcement learning to learn the best paths, speeds, and braking actions in various traffic conditions. The system receives positive reinforcement for safe driving and negative reinforcement for accidents, learning to become a better driver over time.
4. Semi-Supervised Learning
Description:
his approach uses a small amount of labeled data along with a large amount of unlabeled data during the training process.
Example:
 A medical model might be given a few images with tumors labeled by radiologists and then a much larger set of unlabeled medical images to learn from.


Question 3: Define overfitting, underfitting, and the bias-variance tradeoff in machine
learning.

--> Overfitting:
Overfitting occurs when a machine learning model learns the training data too well, including the noise and random fluctuations present in that specific dataset. This leads to excellent performance on the training data but poor performance on new, unseen data, as the model has essentially "memorized" the training examples rather than learning generalizable patterns. Overfit models are characterized by high variance.
Underfitting:
Underfitting occurs when a machine learning model is too simplistic and fails to capture the underlying patterns and relationships within the training data. This results in poor performance on both the training data and unseen test data.
Bias-Variance Tradeoff:
The bias-variance tradeoff is a central concept in machine learning that describes the inherent tension between minimizing bias and minimizing variance in a model's error.

Question 4: What are outliers in a dataset, and list three common techniques for
handling them.


--> Outliers in a dataset are data points that significantly deviate from other observations. They are values that are unusually distant from the majority of the data points, appearing as extreme values at either the high or low end of the data distribution.
Three techniques for handling outliers are:
Deletion:
his involves identifying the outlier data points and completely removing them from the dataset.
Transformation:
This technique involves applying a mathematical function to the data to reduce the impact of outliers.
Capping:
This method involves replacing the outlier values with a less extreme value within the dataset, typically a specified percentile or a value within a certain range (e.g., the 5th or 95th percentile).

Question 6: Write a Python program that:
● Creates a synthetic imbalanced dataset with make_classification() from
sklearn.datasets.
● Prints the class distribution.

```python
from sklearn.datasets import make_classification
import pandas as pd
x,y = make_classification(n_samples = 1000, n_features = 5,n_redundant = 0,n_informative = 3,n_classes = 3,random_state = 42)
df = pd.DataFrame(x,columns = ['f1','f2','f3','f4','f5'])
df['target'] = y
print(df['target'].value_counts())
```

Question 7: Implement one-hot encoding using pandas for the following list of colors:
['Red', 'Green', 'Blue', 'Green', 'Red']. Print the resulting dataframe.

```python
import pandas as pd
from sklearn.preprocessing import OneHotEncoder
coder = OneHotEncoder()
df = pd.DataFrame({'colors' :['Red', 'Green', 'Blue', 'Green', 'Red']})
encoded = coder.fit_transform(df[['colors']]).toarray()
encoded_df = pd.DataFrame(encoded,columns=coder.get_feature_names_out())
final = pd.concat([df,encoded_df], axis = 1)
final
```

Question 8: Write a Python script to:
● Generate 1000 samples from a normal distribution.
● Introduce 50 random missing values.
● Fill missing values with the column mean.
● Plot a histogram before and after imputation.

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
np.random.seed(42)
data = np.random.normal(loc =0,scale = 1,size=1000)
df = pd.DataFrame(data,columns = ['values'])
missing_values = np.random.choice(df.index,size=50,replace = False)
df.loc[missing_values,'values'] = np.nan
df_before = df.copy()
mean_value = df['values'].mean()
df['values'].fillna(mean_value,inplace = True)
plt.figure(figsize =(12,5))
plt.subplot(1,2,1)
plt.hist(df_before['values'].dropna(),bins = 30,edgecolor = 'w')
plt.title('before imputation')
plt.xlabel('values')
plt.ylabel('frequency')


plt.subplot(1,2,2)
plt.hist(df['values'],bins = 30, edgecolor = 'w')
plt.title('after imputation')
plt.xlabel('values')
plt.ylabel('frequency')
plt.show()
```
